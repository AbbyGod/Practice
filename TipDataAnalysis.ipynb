```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)
```

```python
# Preview data
print("First 5 rows:\n", df.head())
print("\nLast 5 rows:\n", df.tail())
```

```python
# Dataset info
print("\nDataset shape:", df.shape)
print("\nMissing values per column:\n", df.isnull().sum())
print("\nSummary statistics:\n", df.describe())
```

```python
# 3️ Basic analysis

# Fraction of males
male_fraction = (df['sex'] == 'Male').mean()
print(f"\nFraction of male customers: {male_fraction:.2f}")

# Average tip overall
print(f"\nAverage tip overall: {df['tip'].mean():.2f}")

# Average tip by day
print("\nAverage tip by day:\n", df.groupby('day')['tip'].mean())

# Average tip by sex
print("\nAverage tip by sex:\n", df.groupby('sex')['tip'].mean())

# Average tip by time (Lunch/Dinner)
print("\nAverage tip by time:\n", df.groupby('time')['tip'].mean())
```

```python
# 4️ Visualizations

# 4a. Bar plot: average tip by day
plt.figure(figsize=(6,4))
sns.barplot(x='day', y='tip', data=df)
plt.title("Average Tip by Day")
plt.show()
```

```python
# 4b. Bar plot: average tip by sex
plt.figure(figsize=(6,4))
sns.barplot(x='sex', y='tip', data=df)
plt.title("Average Tip by Sex")
plt.show()
```

```python
# 4c. Scatter plot: total_bill vs tip
plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', data=df)
plt.title("Tip vs Total Bill (Colored by Sex, Styled by Smoker)")
plt.show()
```

```python
# 4d. Histogram: distribution of tips
plt.figure(figsize=(6,4))
sns.histplot(df['tip'], bins=10, kde=False)
plt.title("Distribution of Tips")
plt.xlabel("Tip Amount")
plt.ylabel("Count")
plt.show()
```

```python
# 4e. Box plot: tips by day (shows spread and outliers)
plt.figure(figsize=(6,4))
sns.boxplot(x='day', y='tip', data=df)
plt.title("Tip Distribution by Day")
plt.show()
```
